```r
can_lang <- read_csv('data/can_lang.csv')
can_lang
```

```r
### filter 
# filter: self-explanatory
aboriginal_lang <- filter(can_lang, category == 'Aboriginal languages')
aboriginal_lang
```

```r
### select 
# select: selects columns
selected_aboriginal_lang <- select(aboriginal_lang, language, mother_tongue)
selected_aboriginal_lang
```

```r
### arrange
# arrange: sorts by selected column
arranged <- arrange(selected_aboriginal_lang, by = desc(mother_tongue)) # desc = descending; default = ascending
arranged
```

```r
### slice
# slice: selects rows
top_ten_lang <- slice(arranged, 1:10) # note: NOT 0-indexed
top_ten_lang
```

```r
### mutate
# mutate: add/modify columns

# NOTE: textbook uses = as an assignment. figure out the difference between <- and = for assignment
population <- 35151728
top_ten_lang_percent <- mutate(top_ten_lang, mother_tongue_percent = mother_tongue / population * 100)
top_ten_lang_percent
```
